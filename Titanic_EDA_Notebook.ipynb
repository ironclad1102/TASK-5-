```python
import seaborn as sns
import pandas as pd

# Load Titanic dataset
df = sns.load_dataset('titanic')

# Show the first 5 rows
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pairplot
sns.pairplot(df.dropna(), hue='sex')
plt.show()

# Heatmap of only numeric columns
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()


# Histogram
df['age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Boxplot
sns.boxplot(x='sex', y='age', data=df)
plt.show()

# Scatterplot
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.show()
```

```python
**Observation:**
- Most passengers are between 20 and 30 years old.
- Very few passengers are older than 60.
- There are some missing values in the age column.

    **Observation:**
- On average, females are slightly younger than males.
- Some male passengers are older than 70.
- The age range for both sexes is wide, from babies to elderly.

    **Observation:**
- Passengers who paid higher fares mostly survived.
- Most passengers paid less than $100.
- There is no strong pattern between age and fare directly.

**Observation:**
- The pairplot shows different variable relationships for males and females.
- Males and females are spread out similarly in terms of age and fare.
- The separation helps us see if sex is related to survival or other features.


**Observation:**
- Fare and Pclass have a strong negative correlation (higher class = lower number, higher fare).
- Age and survival have very little correlation.
- SibSp and Parch have a weak positive correlation, which makes sense as families often travel together.


```


##  Summary of Findings

- Most passengers were young adults between 20 and 30 years old.
- Female passengers had a higher survival rate than males.
- Passengers who paid higher fares were more likely to survive.
- There is a negative correlation between passenger class (Pclass) and fare.
- There was a wide range of ages across both sexes, with some elderly passengers too.
- Most passengers traveled in the 3rd class.
- The data contains some missing values in the `age`, `embark_town`, and `deck` columns.
